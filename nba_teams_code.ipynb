```python
# Gerekli kütüphaneleri import et
from nba_api.stats.endpoints import teamgamelog, leaguegamefinder
from nba_api.stats.static import teams
import pandas as pd
from datetime import datetime
import time
pd.set_option('display.max_columns', 500)
```

```python
# NBA'deki tüm takımları al
all_teams = teams.get_teams()
```

```python
# Son 10 yılı hesapla
end_year = datetime.now().year
start_year = end_year - 10
```

```python
# Her takım için döngü
games = pd.DataFrame()
for team in all_teams:
    print(f"\n{team['full_name']} için veri toplanıyor...")
    team_games_list = []
    
    # Her sezon için döngü
    for season in range(start_year, end_year + 1):
        season_str = f"{season}-{str(season + 1)[-2:]}"
        print(f"{season_str} sezonu işleniyor...")
        
        try:
            # API rate limiting için bekle
            
            # Takımın sezon maçlarını al
            gamelog = teamgamelog.TeamGameLog(
                team_id=team['id'],
                season=season_str
            )
            games_df = gamelog.get_data_frames()[0]
            
            # Sezon bilgisini ekle
            games_df['SEASON'] = season_str
            team_games_list.append(games_df)
            
            print(f"- {len(games_df)} maç bulundu")
            
        except Exception as e:
            print(f"Hata: {season_str} sezonu verisi alınamadı - {str(e)}")
        games = pd.concat([games,games_df],axis =0)
```
